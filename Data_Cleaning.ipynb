```python
import os                               # Operating System
import matplotlib.pyplot as plt         # Graphical ___ data graphs
import seaborn as sns                   # data ---- presents data in graphical format
import pandas as pd                     # reads data
import numpy as np                      # number formation of data
from collections import OrderedDict     # data import for organize form
                ##################################################
from sklearn import datasets            # learning liberary having many datasets ---



```

```python
df = pd.read_csv("Customer Call List.csv")
df
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns = "Not_Useful_Column")
```

```python
# strip = remove the given signs
# in strip method, no "inplace = True" works
df["Last_Name"] = df["Last_Name"].str.strip("._/") # removing all special characters
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]', "")
df['Phone_Number']
```

```python
# converting all phone numbers into strings to apply multiple functions
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number']
```

```python
# putting dashes after 3 characters
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + "-"+ x[3:6] + "-" + x[6:10])
df['Phone_Number']
```

```python
# replacing Nan values to null
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', " ")
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('--', " ")
```

```python
# Spliting address into street, state and zipcode:


df
```

```python
# if a function applies on data frame, no str need
df = df.replace('Yes', 'Y')
df = df.replace('No', 'N')
df
```

```python
# fill Not a value with empty strings
df = df.fillna('')
df
```

```python
# we have to give them a list, which we can call

# for x in df.index:
#   if df.loc[x,"Do_Not_Contact"] == "Y" :
#     df.drop(x, inplace = True)

df = df[df["Do_Not_Contact"] != "Y"]

# another way to drop null values
# df =  df.dropna(subset = "Do_Not_Contact")
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace('', 'N')
df["Do_Not_Contact"]
```

```python
df = df[df["Phone_Number"] != " "]
df
```

```python
# as we assigning it to df, so no need of inplace = True
df = df.drop('Address' , axis = 1)
df
```

```python
# here reseting the index with not saving old one
df.reset_index( drop = True)
```

```python
df = df.drop("index")
```
